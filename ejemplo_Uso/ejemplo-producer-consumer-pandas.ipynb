```python
from confluent_kafka import Producer, Consumer
import random
import pandas as pd
```

Productor

```python
p = Producer({'bootstrap.servers': 'localhost:9092'})

for i in range(100):
    data_producer = {'num_aleatorio':random.randint(1,10)}
    p.produce('testtopic', str(data_producer))
p.flush()
```

Consumidor

```python
c = Consumer({'bootstrap.servers': 'localhost:9092', 'group.id':'migrupo', 'auto.offset.reset': 'earliest'})
c.subscribe(['testtopic'])

data_consumer = []

try:
    while True:
        msg = c.poll(timeout=1.0)
        if msg is None: 
                print('No hay ningún mensaje')
                continue

        """print(f"Valor Mensaje: {msg.value()}")
        print(f"Clave: {msg.key()}")
        print(f"Partición: {msg.partition()}")
        print(f"Offset: {msg.offset()}")"""
        data_consumer.append(msg.value())
except KeyboardInterrupt:
    pass
finally:
        # Close down consumer to commit final offsets.
        c.close()
```

Mostrar DataFreame

```python
df = pd.DataFrame(data_consumer)
print(df.describe())
print(df)
```
